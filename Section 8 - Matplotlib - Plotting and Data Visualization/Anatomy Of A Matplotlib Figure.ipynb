## Matplotlib example workflow

### 0. Import matplotlib and get it ready for plotting in Jupyter

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

### 1. Prepare data

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
```

### 2. Setup plot (where we use object-oriented)

- `(figsize)` - Width and height dimensions for the figure.

```python
# 2. Create figure and axes.
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save and show (saving the whole picture)
fig.savefig("resources/sample-plot.png")
```
