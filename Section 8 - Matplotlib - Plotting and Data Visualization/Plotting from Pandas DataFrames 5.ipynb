```python
%autosave 1
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

heart_disease = pd.read_csv("resources/heart-disease.csv")
heart_disease
```

# Plotting from pandas DataFrames

## Why one should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, it's alright to use the pyplot method.
* When plotting something more advanced, use the OO method.

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

### Pyplot method

```python
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target"); ## c = colour
```

### OO method

* To use this method while in pandas, you have to pass the axis as a parameter.

```python
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);

# Axis is where we have data, meanwhile figure is the entire board.
# The method belows changes the "x" axis limitation.
# ax.set_xlim([45, 100]);
```
