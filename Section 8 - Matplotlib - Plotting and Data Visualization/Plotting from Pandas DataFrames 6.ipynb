```python
%autosave 1
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

heart_disease = pd.read_csv("resources/heart-disease.csv")
over_50 = heart_disease[heart_disease["age"] > 50]
heart_disease
```

## OO method from scratch and using only matplotlib

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholestorol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

# Add a legend (the "*" means to unpack all the values)
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(y=over_50["chol"].mean(), linestyle="--");
```
