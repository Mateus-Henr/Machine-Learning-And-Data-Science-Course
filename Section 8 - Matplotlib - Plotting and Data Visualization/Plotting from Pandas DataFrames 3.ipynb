# Plotting from pandas DataFrames

## Bar graph

```python
x = np.random.rand(10, 4)

df = pd.DataFrame(x, columns=['a', 'b', 'c', 'd']);
df
```

```python
df.plot.bar();
```

```python
df.plot(kind="bar");
```

```python
car_sales.plot(x="Make", y="Odometer (KM)",kind="bar");
```

## Histograms

- The advantage of using the method itself is that you can specify the number of columns.
- A recommended value for bins is one that display something that reminds a normal distribution.

```python
car_sales["Odometer (KM)"].hist();
```

```python
car_sales["Odometer (KM)"].plot(kind="hist");
```

```python
car_sales["Odometer (KM)"].hist(bins=15);
```
