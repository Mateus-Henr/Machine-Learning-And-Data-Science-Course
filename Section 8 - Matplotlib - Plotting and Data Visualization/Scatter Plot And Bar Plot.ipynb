```python
%autosave 1
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Making figures with NumPy arrays

- Line plot
- Scatter plot
- Bar plot
- Histogram
- Subplot

## Create some data

```python
# This function returns evenly spaced numbers over a specified interval.
x = np.linspace(0, 10, 100)
x
```

## Plot the data

### Line Plot

```python
fig, ax = plt.subplots()
ax.plot(x, x ** 2);
```

### Scatter Plot (dots)

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

### Bar Plot

```python
# Plot from a dictionary

nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title="Matt's Nut Butter Store",
      ylabel="Price ($)");
```
