```python
%autosave 1
import pandas as pd
car_sales = pd.read_csv("resources\car-sales.csv")
```

# Creating data

## Creating column from series

```python
seats_column = pd.Series([5, 5, 5, 5, 5])
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
car_sales
```

## Creating column from list

> **Note:** Creating a column from a list you have to provide all the values.

```python
fuel_economy = [7.5, 9.2, 5.3, 4.4, 6.5, 7.5, 9.2, 5.3, 4.4, 6.5]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used (L)"] = (car_sales["Odometer (KM)"] / 100) * car_sales["Fuel per 100KM"]
car_sales
```

## Creating a column from a single value

```python
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["Passed road safety"] = True
car_sales
```

## Removing a column

The function below requires an axis to remove, 1 corresponds to columns.

```python
car_sales = car_sales.drop("Passed road safety", axis=1)
car_sales
```

```python
car_sales["Type"] = "Manual"
car_sales
```
