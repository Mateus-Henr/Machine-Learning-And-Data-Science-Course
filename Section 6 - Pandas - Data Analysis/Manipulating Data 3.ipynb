```python
%autosave 1
import pandas as pd
car_sales = pd.read_csv("resources\car-sales.csv")
```

# Randomizing indexes

- **"frac":** Represents how many percents of the data you want to use.
- It changes the order that we view the data.

```python
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

### Working on 20% of the data

```python
car_sales_shuffled.sample(frac=0.2)
```

## Resetting index order

```python
car_sales_shuffled.reset_index(drop=True, inplace=True)
car_sales_shuffled
```

## Applying a function to a column

```python
# Transforming KM to Miles
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x / 1.6)
car_sales
```
