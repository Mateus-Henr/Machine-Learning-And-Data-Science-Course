```python
%autosave 1
import pandas as pd
car_sales = pd.read_csv("resources\car-sales.csv")
```

# Manipulating Data

### Making strings lower case

```python
car_sales["Make"] = car_sales["Make"].str.lower()
car_sales
```

## Missing Data

Pandas describes missing data as "NaN".

```python
car_sales_missing = pd.read_csv("resources/car-sales-missing-data.csv")
car_sales_missing
```

## Filling missing data

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

## Inplace parameter

This parameter is useful since it removes the need of having to reassign the new value to the variable being modified upon.

```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace=True)
car_sales_missing
```

## Removing rows that have missing values

```python
car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
car_sales_missing_dropped.to_csv("resources/car_sales_missing_dropped.csv")
```
